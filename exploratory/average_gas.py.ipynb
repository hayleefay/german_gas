```python
# split into train and test sets
# now order needs to be by day rather than by station -- reorder by num_days
# num_days is 16 and 15 is station
df.sort_values(by=['num_days', 'station'], inplace=True)
day_30 = df['num_days'].unique()[-30]
train = df.loc[df['num_days'] < day_30]
test = df.loc[df['num_days'] >= day_30]
```

```python
train_ave = train.groupby('station').mean()
test_ave = test.groupby('station').mean()
whole_ave = test.groupby('station').mean()
```

```python
train_ave = train_ave[['e5gas', 'latitude', 'longitude']]
test_ave = train_ave[['e5gas', 'latitude', 'longitude']]
whole_ave = train_ave[['e5gas', 'latitude', 'longitude']]
```

```python
train_ave.to_csv('./../data/train_ave.csv')
test_ave.to_csv('./../data/test_ave.csv')
whole_ave.to_csv('./../data/whole_ave.csv')
```
